```python
def gray_scale(k: int, i: int):
    f = 1
    if i == 0:
        x = 255
    else:
        x = int(255.0 * (k-i) / (k) * f)
    return (x,) # (x,x,x)

for i in range(5):
    print(gray_scale(4, i))
```

```python
def grid_image(n: int, matrix, scale = 20, linesize = 2):
    width = scale * n + linesize
    height = scale * n + linesize
    img = []
    for y in range(height):
        row = ()
        for x in range(width):
            if x % scale < linesize or y % scale < linesize:
                row = row + gray_scale(4, 4)
            else: 
                row = row + gray_scale(4, matrix[(y//scale, x//scale)])
        img.append(row)
    return (width, height, img)

def switch_to_matrix(switch):
    matrix = defaultdict(int)
    for node, targets in switch.items():
        for phase, target in enumerate(targets):
            matrix[(node,target)] = phase + 1
    return matrix
```

# Analyzing rotor switch matchings

```python
def is_complete_graph(switches, phases=4, nodes=16):
    seen = set()
    for phase in range(phases):
        for in_node in range(nodes):
            for i, switch in enumerate(switches):
                out_node = switch[in_node][phase]
                if (in_node, out_node) in seen:
                    return seen, (phase, i, in_node, out_node)
                seen.add((in_node, out_node))
    return True
(is_complete_graph(original_switches), is_complete_graph(switches))
```

```python
def next_next_hops(switches, node, phase, phases=4):
    next_phase = (phase + 1) % phases
    return {sw2[sw1[node][phase]][next_phase] for sw1 in switches for sw2 in switches}
def count_next_next_hops(switches, phases=4, nodes=16):
    return sum(len(set(next_next_hops(switches, node, phase, phases=phases))) for node in range(nodes) for phase in range(phases))
# perms = [r1_perm, r2_perm, r3_perm, r4_perm]
for node in range(16):
    for phase in range(4):
        print(f'{node}, P{phase}: {sorted(next_next_hops(switches, node, phase))}')
count_next_next_hops(original_switches)
```

# Searching for the matchings in Figure 5 (b)
### The following are various attempts of understanding the structure and symmetries of the rotor switch matchings from (Mellette 2024: Realizing RotorNet) leading to a constrained search for permutations of those matchings, which results in larger sets of next-next-hops - eventually finding the insight for the matchings presented in Figure 5 (b) in our paper.
Note: Some of these searches do not terminate quickly and needs to be manually stopped. I leave them here as a documentation of my process of exploration and understanding of the structures and symmetry.

```python
combinable_rows = [(i,j,k,l) for (i,j,k), quadruples_i_j_k in quadruples.items() for l in quadruples_i_j_k.keys() if k < l]
combinable_rows
```

```python
row_perms = sorted([perm for row in combinable_rows for perm in itertools.permutations(row)])
row_perms_dict = {i: [row[1:] for row in row_perms if row[0] == i] for i in range(16)}
# Avoid symmetries from reordering rotor indices by fixing ascending order on first row.
row_perms_dict[0] = [triple for triple in row_perms_dict[0] if triple[0] < triple[1] and triple[1] < triple[2]]
row_perms_dict
```

```python
def push_choice(choice, rotors):
    for i in range(3):
        rotors[i].append(choice[i])
def pop_choice(rotors):
    for i in range(3):
        assert(i < len(rotors) and len(rotors[i]) > 0)
        rotors[i].pop()
def is_safe_choice(choice, rotors):
    for r in range(3):
        if choice[r] in rotors[r]:
            return False
    return True
def search_rec(i, rotors):
    if i == 16:
        yield rotors
    else:
        for choice_i in row_perms_dict[i]:
            if is_safe_choice(choice_i, rotors):
                push_choice(choice_i, rotors)
                yield from search_rec(i+1, rotors)
                pop_choice(rotors)
def search():
    rotors = [[], [], []]
    yield from search_rec(0, rotors)
```

```python
def get_large_union_nnhs(limit, nhs: set[int], nnhs1: list[set[int]], nnhs2: List[Set], nnhs3: List[Set], nnhs4: List[Set]):
    for nnh1 in nnhs1:
        nhnnh1 = nhs.union(nnh1)
        for nnh2 in nnhs2:
            nnh12 = nhnnh1.union(nnh2)
            for nnh3 in nnhs3:
                nnh123 = nnh12.union(nnh3)
                for nnh4 in nnhs4:
                    nnh1234 = nnh123.union(nnh4)
                    if len(nnh1234) >= limit:
                        yield (nhs, nnh1, nnh2, nnh3, nnh4)
def get_disjoint_nnhs(nnhs1: list[set[int]], nnhs2: List[Set], nnhs3: List[Set], nnhs4: List[Set]):
    for nnh1 in nnhs1:
        for nnh2 in nnhs2:
            # if nnh1.isdisjoint(nnh2):
            nnh12 = nnh1.union(nnh2)
            if len(nnh12) >= 4:
                for nnh3 in nnhs3:
                    # if nnh12.isdisjoint(nnh3):
                    nnh123 = nnh12.union(nnh3)
                    if len(nnh123) >= 8:
                        for nnh4 in nnhs4:
                            # if nnh123.isdisjoint(nnh4):
                            nnh1234 = nnh123.union(nnh4)
                            if len(nnh1234) >= 12:
                                yield (nnh1, nnh2, nnh3, nnh4)
```

```python
def search_disjoint(phase = 0, node = 0):
    for choice in range(len(row_perms_dict[node])):
        nhs = [r1[n][phase] for n in get_stuff(node, choice)]
        yield from get_disjoint_nnhs(get_nnhs(nhs[0], next_phase(phase)), get_nnhs(nhs[1], next_phase(phase)), get_nnhs(nhs[2], next_phase(phase)), get_nnhs(nhs[3], next_phase(phase)))

def search_large_union(phase = 0, node = 0):
    for limit in range(16, 0, -1):
        stop = False
        for choice in range(len(row_perms_dict[node])):
            nhs = [r1[n][phase] for n in get_stuff(node, choice)]
            if sum(1 for _ in get_large_union_nnhs(limit, set(nhs), get_nnhs(nhs[0], next_phase(phase)), get_nnhs(nhs[1], next_phase(phase)), get_nnhs(nhs[2], next_phase(phase)), get_nnhs(nhs[3], next_phase(phase)))) > 0:
                stop = True
                yield from get_large_union_nnhs(limit, set(nhs), get_nnhs(nhs[0], next_phase(phase)), get_nnhs(nhs[1], next_phase(phase)), get_nnhs(nhs[2], next_phase(phase)), get_nnhs(nhs[3], next_phase(phase)))
        if stop:
            print("Limit:", limit)
            break
def search_large_union2(phase = 0, node = 0):
    for choice in range(len(row_perms_dict[node])):
        nhs = [r1[n][phase] for n in get_stuff(node, choice)]
        if sum(1 for _ in get_large_union_nnhs(16, set(nhs+[node]), get_nnhs(nhs[0], next_phase(phase)), get_nnhs(nhs[1], next_phase(phase)), get_nnhs(nhs[2], next_phase(phase)), get_nnhs(nhs[3], next_phase(phase)))) > 0:
            yield get_stuff(node, choice)
        # for x in get_large_union_nnhs(16, set(nhs), get_nnhs(nhs[0], next_phase(phase)), get_nnhs(nhs[1], next_phase(phase)), get_nnhs(nhs[2], next_phase(phase)), get_nnhs(nhs[3], next_phase(phase))):
            # yield (get_stuff(node, choice), x)
def good_options_(phase, node, options_per_row):
    for option in options_per_row[node]:
        nhs = [r1[n][phase] for n in (node,) + option]
        reachable = set(nhs+[node])
        def get_nnhs(nh, phase):
            return to_set_list({tuple(sorted(r1[n][phase] for n in (nh,) + o)) for o in options_per_row[nh]})
        do_break = False
        for nnh1 in get_nnhs(nhs[0], next_phase(phase)):
            reachable1 = reachable.union(nnh1)
            for nnh2 in get_nnhs(nhs[1], next_phase(phase)):
                reachable2 = reachable1.union(nnh2)
                for nnh3 in get_nnhs(nhs[2], next_phase(phase)):
                    reachable3 = reachable2.union(nnh3)
                    for nnh4 in get_nnhs(nhs[3], next_phase(phase)):
                        reachable4 = reachable3.union(nnh4)
                        if len(reachable4) >= 16:
                            yield option
                            do_break = True
                            break
                    if do_break:
                        break
                if do_break:
                    break
            if do_break:
                break

def intersect_all(sets: list[set]):
    inters = sets[0]
    for set in sets[1:]:
        inters.intersection_update(set)
    return inters
def good_options(node, options_per_row):
    return sorted(intersect_all([set(d for d in good_options_(phase, node, options_per_row)) for phase in range(4)]))

```

```python
def smart_search_rec(node, rotors, options_per_row):
    if node == 16:
        yield rotors
    else:
        new_options_per_row = options_per_row.copy()
        for choice in good_options(node, options_per_row):
            if is_safe_choice(choice, rotors):
                # nhs = [r1[n][phase] for n in (node,) + choice_i]
                # def get_nnhs(nh, phase, perm=r1_perm):
                    # return to_set_list({tuple(sorted(r1[perm[n]][phase] for n in get_stuff(nh, c))) for c in range(len(row_perms_dict[nh]))})
                # if sum(1 for _ in get_large_union_nnhs(16, set(nhs+[node]), get_nnhs(nhs[0], next_phase(phase)), get_nnhs(nhs[1], next_phase(phase)), get_nnhs(nhs[2], next_phase(phase)), get_nnhs(nhs[3], next_phase(phase)))) > 0:
                push_choice(choice, rotors)
                new_options_per_row[node] = [choice]
                yield from smart_search_rec(node+1, rotors, new_options_per_row)
                pop_choice(rotors)
def smart_search():
    rotors = [[], [], []]
    yield from smart_search_rec(0, rotors, row_perms_dict)
```
